8 find the sum of each row of matrix of size m x n. For example for the following matrix output will be like this  
2 11 7 12  
5 2 9 15  
8 3 10 42  
sum of row 1 = 32  
sum of row 2 = 31  
sum of row 3 = 63

```python
# read dimensions
m = int(input("Enter number of rows (m): "))
n = int(input("Enter number of columns (n): "))

row_sums = []
for i in range(m):
    row = list(map(int, input(f"Enter row {i+1} (space-separated {n} ints): ").split()))
    if len(row) != n:
        raise ValueError(f"Expected {n} integers, got {len(row)}.")
    row_sums.append(sum(row))


for i, s in enumerate(row_sums, 1):
    print(f"sum of row{i}={s}")  
```

9. Write a program to add two matrices of size n × m.

```python
# add_matrices` with validation, interactive/sample-run, 

def add_matrices(A, B):
    """Return element-wise sum of matrices A and B.

    A and B should be non-empty lists of lists with matching dimensions.
    Raises ValueError for invalid inputs or mismatched dimensions.
    """
    if not isinstance(A, list) or not isinstance(B, list):
        raise ValueError("A and B must be matrices (list of lists)")
    if len(A) != len(B):
        raise ValueError("Matrices must have the same number of rows")
    if len(A) == 0:
        return []

    # Ensure each row is the same length and rows are numeric
    m = len(A[0])
    for row in A:
        if not isinstance(row, list) or len(row) != m:
            raise ValueError("All rows in A must have the same length")
    for row in B:
        if not isinstance(row, list) or len(row) != m:
            raise ValueError("All rows in B must have the same length as A")

    C = [[A[i][j] + B[i][j] for j in range(m)] for i in range(len(A))]
    return C


def read_matrix(n, m, label=""):
    rows = []
    for i in range(n):
        row = list(map(int, input(f"Enter row {i+1} of {label} (space-separated {m} ints): ").split()))
        if len(row) != m:
            raise ValueError(f"Expected {m} integers, got {len(row)}.")
        rows.append(row)
    return rows


# Interactive / sample-run behavior (press Enter to run sample)
try:
    first = input("Enter n m (or press Enter to run sample): ").strip()
except EOFError:
    first = ""

if not first:
    sample = "2 3\n1 2 3\n4 5 6\n7 8 9\n10 11 12\n"
    print("No input detected — running sample input:\n" + sample)
    it = iter(sample.split())
    n = int(next(it)); m = int(next(it))
    A = [[int(next(it)) for _ in range(m)] for _ in range(n)]
    B = [[int(next(it)) for _ in range(m)] for _ in range(n)]
else:
    n, m = map(int, first.split())
    A = read_matrix(n, m, label="A")
    B = read_matrix(n, m, label="B")

# Compute and print result
C = add_matrices(A, B)
print("Result matrix:")
for row in C:
    print(*row)

# Quick self-tests
assert add_matrices([[1]], [[2]]) == [[3]]
assert add_matrices([[1,2],[3,4]], [[4,3],[2,1]]) == [[5,5],[5,5]]
print("Tests passed ✅")
```

10. write a prog to multiply tow matrices.

```python
# tow matrices to be define
a = [[1, 2, 3], [4, 5, 6]]
b = [[7, 8,], [9, 10], [11, 12]]
# resultmatrix (2x2)
result = [[0, 0], [0, 0]]
# logic of multiplication (to ch eck three loops)
for i in range(len(a)):
    for j in range(len(b[0])):
        for k in range(len(b)):
            result[i][j] += a[i][k] * b[k][j]
# for print final result
print("final result matrix:")  
for r in result:
    print(r)
```
